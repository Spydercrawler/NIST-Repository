```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
%matplotlib notebook
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
```

```python
x_vals = np.linspace(0,10,50)
sin_vals = np.sin(x_vals)
x_vals,sin_vals = np.meshgrid(x_vals,sin_vals)
random_vals = [np.linspace(10,0,50).tolist()] * 50
print random_vals[0]
```

```python
plt.ion()
surf = ax.plot_surface(x_vals,sin_vals,random_vals, cmap=cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.draw()
```
