```python
import itertools

class depvarcomb:
    dependentvars = []
    def add_dependent(self,name,values):
        self.dependentvars.append([name,values])
    def clear_dependents(self):
        self.dependentvars = []
    def add_basic_dependents(self):
        self.add_dependent("xpoints",range(11))
        self.add_dependent("connection type",["HDMI","DVI","Ethernet","USB","Micro USB"])
        self.add_dependent("device type",["Television","Computer","Game Console"])
    def print_combinations(self):
        vallst = [lst[1] for lst in self.dependentvars]
        for lst in list(itertools.product(*vallst)):
            print(lst)
    def print_dependent_list(self):
        for lst in self.dependentvars:
            print(lst[0] + ": " + str(lst[1]))
```

```python
dvc = depvarcomb()
```

```python
dvc.add_basic_dependents()
dvc.print_combinations()
```

```python
dvc.add_dependent("I Dont Even Know",["Some","Sample","Text"])
dvc.print_combinations()
```

```python
dvc.print_dependent_list()
```

```python
dvc.clear_dependents()
dvc.print_dependent_list()
dvc.print_combinations()
dvc.add_dependent("Pi Digits",[3,1,4,1,5,9,2,6,5,3,5,9])
dvc.print_combinations()
dvc.print_dependent_list()
```
