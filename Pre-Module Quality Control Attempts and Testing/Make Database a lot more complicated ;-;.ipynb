```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
```

```python
db = joblib.load('cries_internally.pkl')
```

```python
ndb = ndb = [[j[0],None] for j in db]
path='Combined_Two_Port_Check_Standard.csv'
df=pd.read_csv(path)
system_letters = df["System_Letter"].unique()
```

```python
def get_dataset(lst,start,make_arr=True):
    date = lst[start][14]
    dataset_still_going=True
    i=start+1
    tdb = [lst[start]]
    while dataset_still_going:
        if i>=len(lst) or lst[i][14] != date:
            dataset_still_going = False
        else:
            tdb.append(lst[i])
            i += 1
    if make_arr:
        return np.array(tdb)
    else:
        return tdb
def get_dataset_end(lst,start):
    date = lst[start][14]
    dataset_still_going=True
    i=start+1
    while dataset_still_going:
        if i>=len(lst) or lst[i][14] != date:
            dataset_still_going = False
        else:
            i += 1
    return i
```

```python
for i in range(len(db)):
    arraddon = [[j,[]] for j in system_letters]
    for j in range(len(db[i][1])):
        k=0
        while k<len(db[i][1][j][1]):
            arraddon[j][1].append(get_dataset(db[i][1][j][1],k))
            k=get_dataset_end(db[i][1][j][1],k)
    ndb[i][1] = arraddon
```
