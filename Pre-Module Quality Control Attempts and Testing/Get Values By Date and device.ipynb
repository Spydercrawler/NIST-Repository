```python
import pandas as pd
import numpy as np
```

```python
path='Combined_Two_Port_Check_Standard.csv'
df=pd.read_csv(path)
```

```python
device_ids = df["Device_Id"].unique()
idlst = [[id, None] for id in device_ids]
print idlst
```

```python
still_going=True
i=0
while still_going:
    if i>=df.shape[0]:
        still_going=False
    else:
        for j in range(len(idlst)):
            if df.iloc[i]["Device_Id"] == idlst[j][0]:
                if idlst[j][1] is None:
                    idlst[j][1]=df[i:i+1]
                    break;
                else:
                    idlst[j][1] = idlst[j][1].append(df[i:i+1])
                    break;
    if i%10000==0:
        print "i is now: " + str(i)
    i+=1
print idlst
```

```python
from sklearn.externals import joblib
joblib.dump(idlst,'Combined_Two_Port_Check_Standard_by_dev_id.pkl')
```
