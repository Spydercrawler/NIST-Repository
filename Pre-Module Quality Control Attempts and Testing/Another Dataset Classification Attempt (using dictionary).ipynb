```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv("Combined_Two_Port_Check_Standard.csv")
```

```python
devices=data_frame["Device_Id"].unique()
print devices
```

```python
device_data_dictionary={}
for device_id in devices:
    df=data_frame[data_frame["Device_Id"]==device_id]
    device_data_dictionary[device_id]=df
```

```python
device_data_dictionary["CTN210"]
```

```python
better_dictionary={}
for key in device_data_dictionary.keys():
    systems = device_data_dictionary[key]["System_Id"].unique()
    system_dictionary={}
    for system in systems:
        df = device_data_dictionary[key][device_data_dictionary[key]["System_Id"]==system]
        system_dictionary[system] = df
    better_dictionary[key] = system_dictionary
```

```python
better_dictionary["CTN210"]['System 2,7']
```

```python
final_dictionary = {}
for dkey in better_dictionary.keys():
    final_dictionary[dkey] = {}
    for skey in better_dictionary[dkey].keys():
        device_system_dictionary = {}
        measurement_dates = better_dictionary[dkey][skey]["Measurement_Date"].unique()
        for date in measurement_dates:
            df=better_dictionary[dkey][skey][better_dictionary[dkey][skey]["Measurement_Date"]==date]
            device_system_dictionary[date] = df
        final_dictionary[dkey][skey] = device_system_dictionary
```

```python
final_dictionary["CTN210"]['System 2,7'].keys()
```

```python
import matplotlib.pyplot as plt
plt.figure()
for k in final_dictionary["CTN210"]['System 2,7'].keys():
    plt.plot(final_dictionary["CTN210"]['System 2,7'][k]["Frequency"].values.tolist(),
             final_dictionary["CTN210"]['System 2,7'][k]["magS11"].values.tolist())
plt.show()
```

```python
from sklearn.externals import joblib
joblib.dump(final_dictionary,"Two_Port_Standard_Dictionary.pkl")
```
