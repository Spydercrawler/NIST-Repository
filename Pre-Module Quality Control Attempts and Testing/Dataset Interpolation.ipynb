```python
from sklearn.externals import joblib
import numpy as np
import scipy as sp
from scipy import interpolate
```

```python
datasets = joblib.load("../Combined_Two_Port_Check_Standard_ct.pkl")
```

```python
interpolationfunctions = []
for i in range(len(datasets[0])):
    frequencies = datasets[0][i]
    magS11i = interpolate.interp1d(frequencies,datasets[1][i][0])
    magS21i = interpolate.interp1d(frequencies,datasets[1][i][1])
    magS22i = interpolate.interp1d(frequencies,datasets[1][i][2])
    interpolationfunctions.append([magS11i,magS21i,magS22i])
print interpolationfunctions
```

```python
interpolatedfrequencies = []
interpolatedmags = []
interpolation_length = 50
for i in range(len(interpolationfunctions)):
    functions = interpolationfunctions[i]
    frequencies = np.array(datasets[0][i])
    ifrequencies = np.linspace(frequencies.min(),frequencies.max(),interpolation_length)
    magS11i = functions[0](ifrequencies)
    magS21i = functions[1](ifrequencies)
    magS22i = functions[2](ifrequencies)
    interpolatedfrequencies.append(ifrequencies)
    interpolatedmags.append([magS11i,magS21i,magS22i])
print interpolatedfrequencies[0]
print interpolatedmags[0]
```

```python
interpolateddatasets = np.array([interpolatedfrequencies,interpolatedmags])
joblib.dump(interpolateddatasets,"Combined_Two_Port_Check_Standard_ct_interpolated.pkl")
```
