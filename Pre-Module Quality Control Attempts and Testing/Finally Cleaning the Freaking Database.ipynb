```python
db = joblib.load('Combined_Two_Port_Check_Standard_Made_Complicated.pkl')
```

```python
db2 = list(db)
for i in range(len(db)):
    for j in range(len(db[i][1])):
        for k in range(len(db[i][1][j][1])):
            db2[i][1][j][1][k] = [[l[0],l[3],l[5],l[7]] for l in db[i][1][j][1][k]]
```

```python
data_frame=pandas.read_csv(r"Combined_Two_Port_Check_Standard.csv")
```

```python
devices=data_frame["Device_Id"].unique()
```

```python
print devices
```

```python
device_data_dictionary={}
for device_id in devices:
    df=data_frame[data_frame["Device_Id"]==device_id]
    device_data_dictionary[device_id]=df
```

```python
device_data_dictionary["CTN210"]
```

```python
systems=device_data_dictionary["CTN210"]["System_Id"].unique()
CTN210_system_dictionary={}
for system in systems:
    df=device_data_dictionary["CTN210"][device_data_dictionary["CTN210"]["System_Id"]==system]
    CTN210_system_dictionary[system]=df
```

```python
CTN210_system_dictionary.keys()[0]
```

```python
measurement_dates=CTN210_system_dictionary["System 2,7"]["Measurement_Date"].unique()
CTN210_system27_dictionary={}
for date in measurement_dates:
    df=CTN210_system_dictionary["System 2,7"][CTN210_system_dictionary["System 2,7"]["Measurement_Date"]==date]
    CTN210_system27_dictionary[date]=df
```
