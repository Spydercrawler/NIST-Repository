```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
```

```python
datadict = joblib.load("Two_Port_Standard_Dictionary.pkl")
```

```python
ntdcount = 0
for dkey in datadict.keys():
    for skey in datadict[dkey].keys():
        for tkey in datadict[dkey][skey].keys():
            new_t_dataframe = None
            frequencies = datadict[dkey][skey][tkey]["Frequency"].unique()
            for f in frequencies:
                df = datadict[dkey][skey][tkey][datadict[dkey][skey][tkey]["Frequency"]==f]
                if new_t_dataframe is None:
                    new_t_dataframe = df.iloc[[0]]
                else:
                    new_t_dataframe = pd.concat([new_t_dataframe,df.iloc[[0]]])
            ntdcount += 1
            if new_t_dataframe.shape[0] < 10:
                del datadict[dkey][skey][tkey]
                print "removed short dataframe"
                if ntdcount % 10 == 0:
                    print str(ntdcount) + " dataframes have been updated or removed!"
                continue
            if ntdcount % 10 == 0:
                print str(ntdcount) + " dataframes have been updated or removed!"
            datadict[dkey][skey][tkey] = new_t_dataframe
```

```python
import copy
import scipy as sp
from scipy import interpolate
ddict2 = copy.deepcopy(datadict)
num_values = 50
for dkey in datadict.keys():
    for skey in datadict[dkey].keys():
        fmaxofmins = None
        fminofmaxes = None
        for tkey in datadict[dkey][skey].keys():
            if fmaxofmins is None:
                fmaxofmins = datadict[dkey][skey][tkey]["Frequency"].values.min()
            elif fmaxofmins < datadict[dkey][skey][tkey]["Frequency"].values.min():
                fmaxofmins = datadict[dkey][skey][tkey]["Frequency"].values.min()
            if fminofmaxes is None:
                fminofmaxes = datadict[dkey][skey][tkey]["Frequency"].values.max()
            elif fminofmaxes > datadict[dkey][skey][tkey]["Frequency"].values.max():
                fminofmaxes = datadict[dkey][skey][tkey]["Frequency"].values.max()
        for tkey in datadict[dkey][skey].keys():
            frequencies = datadict[dkey][skey][tkey]["Frequency"].values.tolist()
            mags11s = datadict[dkey][skey][tkey]["magS11"].values.tolist()
            mags21s = datadict[dkey][skey][tkey]["magS21"].values.tolist()
            mags22s = datadict[dkey][skey][tkey]["magS22"].values.tolist()
            mags11func = interpolate.interp1d(frequencies,mags11s,kind="cubic")
            mags21func = interpolate.interp1d(frequencies,mags21s,kind="cubic")
            mags22func = interpolate.interp1d(frequencies,mags22s,kind="cubic")
            nfrequencies = np.linspace(fmaxofmins,fminofmaxes,num_values)
            print dkey + ", " + skey + ", " + tkey
            try:
                nmags11s = mags11func(nfrequencies)
                nmags21s = mags21func(nfrequencies)
                nmags22s = mags22func(nfrequencies)
                ddict2[dkey][skey][tkey] = {"Frequency" : nfrequencies,
                                            "magS11" : nmags11s,
                                            "magS21" : nmags21s,
                                            "magS22" : nmags22s}
            except ValueError:
                print "This one had an issue"
                del datadict[dkey][skey][tkey]
                continue
```
