```python
import pandas as pd
import numpy as np
```

```python
path='Combined_Two_Port_Check_Standard.csv'
df=pd.read_csv(path)
```

```python
def get_dataset(df, start, make_arr=True):
    date = df[start:(start+1)]['Measurement_Date'].values[0]
    df2 = df[start:start+1]
    dataset_still_going=True
    i=start+1
    while dataset_still_going:
        if i>=df.shape[0] or df[i:i+1]['Measurement_Date'].values[0] != date:
            dataset_still_going = False
        else:
            df2 = df2.append(df[i:i+1])
            i+= 1
    if make_arr:
        return df2.values
    else:
        return df2

def get_dataset_end(df, start):
    date = df[start:(start+1)]['Measurement_Date'].values[0]
    dataset_still_going=True
    i=start+1
    while dataset_still_going:
        if i>=df.shape[0] or df[i:i+1]['Measurement_Date'].values[0] != date:
            dataset_still_going = False
        else:
            i += 1
    return i-1
```

```python
datasets=[]
i=0
while i<df.shape[0]:
    datasets.append(get_dataset(df,i).tolist())
    i=get_dataset_end(df,i)+1
datasetsarr = np.array(datasets)
```

```python
from sklearn.externals import joblib
joblib.dump(datasetsarr,"Combined_Two_Port_Check_Standard.pkl")
```
