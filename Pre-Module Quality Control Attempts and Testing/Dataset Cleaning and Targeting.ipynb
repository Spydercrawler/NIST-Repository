```python
from sklearn.externals import joblib
import numpy as np
```

```python
datasets = joblib.load("../Combined_Two_Port_Check_Standard.pkl")
datasetslist = datasets.tolist()
```

```python
datasetslist = [v for v in datasetslist if len(v) >=10]
```

```python
datasetscleaned=np.array(datasetslist)
```

```python
print len(datasets)
print len(datasetscleaned)
```

```python
ctdatasetfrequencies = []
ctdatasetmags = []
for dset in datasetscleaned:
    nsetf = []
    nsetm = [[],[],[]]
    for lst in dset:
        nsetf.append(lst[0])
        nsetm[0].append(lst[3])
        nsetm[1].append(lst[5])
        nsetm[2].append(lst[7])
    ctdatasetfrequencies.append(nsetf)
    ctdatasetmags.append(nsetm)
print ctdatasetfrequencies[0]
print ctdatasetmags[0]
```

```python
datasets_cleaned_targeted=np.array([ctdatasetfrequencies,ctdatasetmags])
```

```python
joblib.dump(datasets_cleaned_targeted,"Combined_Two_Port_Check_Standard_ct.pkl")
```
