# The Multiple_Edits Method
This was a method I made to make more concise code, and it basically does multiple TableDictionarySystem methods at once.

First, to show off the method, we have to the imports and get the actual TableDictionarySystem:

```python
# Imports:
from MachineLearningModule import *
import pandas as pd

# Make TableDictionarySystem:
pandas_dataframe = pd.read_csv("Combined_Two_Port_Check_Standard.csv")
data_table_dictionary_system = split_table_by_column(pandas_dataframe,"Device_Id")
```

The multiple_edits method has keyword arguments as its argument. Each keyword must be the name of another method, and each keyword's value must be a dictionary containing the name of the method's arguments as keys, and the values of the arguments as values.

Here is an example of it being called:

```python
data_table_dictionary_system.multiple_edits(split_by_columns={"columns":("System_Id","Measurement_Date")},
                                       remove_column_duplicates={"column":"Frequency"},
                                       remove_short_tables={"row_count":10},
                                       keep_only_certain_columns={"columns":("Frequency","magS11","magS21","magS22")})
```

Here is what a single dataset's table looks like after all of this:

```python
data_table_dictionary_system['C00001']['HP8510']['9 Jul 2010']
```
